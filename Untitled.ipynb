```python
df_train = pd.read_csv("audioset_train_strong.tsv", sep='\t')
df_eval = pd.read_csv("audioset_eval_strong.tsv", sep='\t')


ont = json.load(open("ontology.json"))
ontology = pd.DataFrame(ont)

def getIds(df,id):
    item = df[df.id==id]
    children = item.child_ids.values[0]
    res = [id]
    if len(children)==0:
        return res
    for ch in children:
        res.extend(getIds(df,ch))
    return res
        
animal_ids = getIds(ontology,'/m/0jbk')
vgg_df = pd.read_csv('vggsound.csv', header=None, names=['segment_id','start_time_seconds','label','split'])


animals_df_train = df_train[df_train.label.isin(animal_ids)]
animals_df_eval = df_eval[df_eval.label.isin(animal_ids)]

id_to_label = {id:ontology[ontology.id==id].name.values[0] for id in set(animal_ids)}

animals_df_train.loc[:,'label_name'] = animals_df_train['label'].apply(lambda id: id_to_label[id])
animals_df_eval.loc[:,'label_name'] = animals_df_eval['label'].apply(lambda id: id_to_label[id])
animals_df_train.loc[:,'len'] = animals_df_train['end_time_seconds'] - animals_df_train['start_time_seconds']
animals_df_eval.loc[:,'len'] = animals_df_eval['end_time_seconds'] - animals_df_eval['start_time_seconds']

animals_df_train = animals_df_train.drop(animals_df_train[animals_df_train.len < 0.3].index)
animals_df_eval = animals_df_eval.drop(animals_df_eval[animals_df_eval.len < 0.3].index)

len(animals_df_train), len(animals_df_eval)
```

```python
class DataExtractor:
    def __init__(self,audio_path, img_path, data, vgg=False):
        self.df = data
        self.audio_path = audio_path
        self.img_path = img_path
        self.vgg = vgg
        self.log = logging.getLogger('data_extractor')
        
    def downloadVideo(self,video_id):
        video_url = f"https://www.youtube.com/watch?v={video_id}"
        stream = YouTube(video_url).streams.get_highest_resolution()
        stream.download(filename=f'{video_id}.mp4')
        
    def extract_data_pair(self,video_id,segments):
        clip = VideoFileClip(f'{video_id}.mp4')
        for start,end in segments:
            subclip = clip.subclip(start, end)
            audio = subclip.audio
            frame = subclip.get_frame((end-start)/2)
            frame = Image.fromarray(frame)
            segment_id = f'{video_id}_{start}_{end}'
            audio.write_audiofile(f'{self.audio_path}/{segment_id}.wav',verbose=False, logger=None)
            frame.save(f'{self.img_path}/{segment_id}.png')
        clip.close()
        return audio,frame

    def extract_data_from_video(self,video_id):
        try:
            self.downloadVideo(video_id)
            sts = self.df[self.df.segment_id==video_id].start_time_seconds.values
            if self.vgg: # vgg dataset only provides start time
                ets = sts + 10
            else:
                ets = self.df[self.df.segment_id==video_id].end_time_seconds.values
            segments = list(zip(sts,ets))
            self.extract_data_pair(video_id,segments)
            os.remove(f'{video_id}.mp4')
        except Exception as err:
            self.log.error(err)
        
    def extract_data(self):
        videos = self.df.segment_id.unique()
        with ProcessPoolExecutor() as pool:
            pool.map(self.extract_data_from_video,videos)
    
        
logging.basicConfig(
    level=logging.ERROR, 
    format='%(process)d:%(levelname)s:%(name)s:%(message)s',
    handlers = [logging.StreamHandler(), logging.FileHandler("logging.log")])

de = DataExtractor('audio', 'img', animals_df_train[300:330])
de.extract_data()
```
